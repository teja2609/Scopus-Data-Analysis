```python
data = ds[['Authors', 'Title', 'Correspondence Address',
        'Source title']]
```

```python
data['index'] = range(len(data))
```

```python
data.isna().sum()
```

```python
features = [['Authors', 'Title', 'Correspondence Address',
        'Source title']]
```

```python
for feature in features:
    data[feature]=data[feature].fillna('')
```

```python
data.isna().sum()
```

```python
def combine_features(row):
    return row['Title']+" "+row['Correspondence Address']+" "+row['Source title']
```

```python
data["Author_Details"]=data.apply(combine_features,axis=1)
```

```python
data['Author_Details'][0]
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
def cleaning(text):
    review = re.sub('[^a-zA-Z]',' ',text)
    return review
        
data['Author_Details'] = data['Author_Details'].apply(cleaning)
```

```python
def lower(review):
    review=review.lower()
    #review=review.split()

    
    return review
data['Author_Details'] = data['Author_Details'].apply(lower)  
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop = set(stopwords.words('english'))
stemmer = PorterStemmer()
```

```python
import nltk
nltk.download('punkt')
```

```python
def lemmatize(text):
    final_text=[]
    wordnet_lemmatizer = WordNetLemmatizer()
    token_words=word_tokenize(text)
    for word in token_words:
        
        if word not in stop:
            word = wordnet_lemmatizer.lemmatize(word)
            final_text.append(word)
    return " ".join(final_text) 
data['Author_Details'] = data['Author_Details'].apply(lemmatize)   
        
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['Author_Details'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(tfidf_matrix)
print(similarity_scores)
```

```python
def get_name_from_index(index):
    names=[]
    a=data[data.index == index]["Authors"]
    for key,value in enumerate(a):
        names.append(value)
    return names

def get_index_from_name(Author):
    indices=[]
    a=data[data.Authors == Author]["index"]
    for key,value in enumerate(a):
        indices.append(value)
    return indices
```

```python
author_index = np.random.randint(0,len(data))
author = get_name_from_index(author_index)
author
```

```python
similar_Author_Scores=list(enumerate(similarity_scores[author_index]))
```

```python
sorted_similar_Author_scores=sorted(similar_Author_Scores,key= lambda x:x[1],reverse=True)
```

```python
sorted_similar_Author_scores
```

```python
co_authors = []
for key,value in enumerate(sorted_similar_Author_scores):
    if value[1]>=0.97:
        co = get_name_from_index(value[0])
        if co != author and co != '[]' :
            co_authors.append(co)
            print('Author {}-----already in Co_Author Relationship--------------> {}'.format(author,co))
    if value[1]>0.5 and value[1]<0.97:
        
        co = get_name_from_index(value[0])
        if co != author and co != '[]' :
            
            co_authors.append(co)
            print('Author {}----- Co_Author Relationship--------------> {}'.format(author,co))
        
    
    
   
    
```
